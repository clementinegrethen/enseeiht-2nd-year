2)D'après les graphiques, lorsque l'on augmente tho, mu augmente également , mais le nombre d'itération nécessaire diminue. 


```julia
using Plots

lambda0 = 2
mu0 = 10
epsilon = 1.
tol = 1e-5
max_iters = 1000
tol_erreur = 1e-4

iters=[]
mus=[]

for tho in [10*i for i in 1:10]
    options = [epsilon, tol, max_iters, lambda0, mu0, tho]
    xmin,fxmin,flag,nbiters, muks, lambdaks = Lagrangien_Augmente("cauchy",fct2,contrainte2,grad_fct2,hess_fct2,grad_contrainte2,hess_contrainte2,pts2.x03,options)
    afficher_resultats("Lagrangien augmenté avec "*"cauchy","fonction 2","x03",xmin,fxmin,flag,sol_fct2_augm,nbiters)
    push!(mus, muks[length(muks)])
    push!(iters, nbiters)
end

print([10*i for i in 1:10])
println(mus)
println(iters)

p2 = plot([10*i for i in 1:10], mus, title="mu en fonction de tho")
p1 = plot([10*i for i in 1:10], iters, title="iters en fonction de tho")

plot(p1,p2,legend=false)
```
